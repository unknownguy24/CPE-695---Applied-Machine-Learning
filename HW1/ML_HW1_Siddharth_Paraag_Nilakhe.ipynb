```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
###################################################################################

#section1
#Please plot the noisy data and the polynomial you found (in the same figure). You can use any 
#value of m selected from 2, 3, 4, 5, 6. 
#Here we choose m = 2

###################################################################################


x = x.flatten()
y = y.flatten()
x1=np.sort(x)
y1=np.sort(y)
poly=np.poly1d(np.polyfit(x1, y1, 2))
plt.plot(x,y,'ro',label = 'Noisy Data')
plt.plot(x1, poly(x1),label = 'Polynomial')
plt.legend()
plt.title('M = 2')
plt.show()
```

```python
###################################################################################

#section2
#Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m.

###################################################################################

from sklearn.metrics import mean_squared_error
m = [1,2,3,4,5,6,7,8]
mse = []
x = x.flatten()
y = y.flatten()
for a in m:
    x1=np.sort(x)
    y1=np.sort(y)
    poly=np.poly1d(np.polyfit(x1, y1, a))
    plt.plot(x,y,'ro',label = 'Noisy Data')
    plt.plot(x1, poly(x1),label = 'Polynomial')
    plt.legend()
    plt.title('m = %i'%a)
    plt.show()
    s = mean_squared_error(y1,poly(x1))
    mse.append(s)
```

```python
###################################################################################

#section 2 continued

###################################################################################

print(mse)
plt.plot(m,mse,'g')
plt.xlabel('Value of m')
plt.ylabel('Value of MSE')
plt.show()
```

```python
#We choose M as 3 (best choice of m)
```

```python
###################################################################################

#section3
#Change variable noise_scale to 200, 300, 400, 600, 800, 1000 respectively, re-run the algorithm 
#and plot the polynomials with the m found in 2). Discuss the impact of noise scale to the 
#accuracy of the returned parameters. [You need to plot a figure like in 1) for EACH choice of 
#noise_scale.] 


###################################################################################

noise = [200,300,400,600,800,1000]
for noise_scale in noise:
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    x = x.flatten()
    y = y.flatten()
    x1=np.sort(x)
    y1=np.sort(y)
    poly=np.poly1d(np.polyfit(x1, y1, 3))
    plt.plot(x,y,'ro')
    plt.plot(x1, poly(x1))
    plt.show()
    
```

```python
#As the noise_scale increases the noisy data gets more dispersed and move further away from the polynomial line 
#that means that the mean squared error will increase with an increase in the noise_scale
```

```python
##############################################################################################################

#Q4.Change variable number_of_samples to 40, 30, 20, 10 respectively, re-ran the algorithm and
#   plot the polynomials with the m found in 2). Discuss the impact of the number of samples to the accuracy
#   of the returned parameters.You need to plot a figure like in 1) for each choice of number_of_samples

##############################################################################################################

samples = [40,30,20,10]
for number_of_samples in samples:
    noise_scale = 100
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    x = x.flatten()
    y = y.flatten()
    x1=np.sort(x)
    y1=np.sort(y)
    poly=np.poly1d(np.polyfit(x1, y1, 3))
    plt.plot(x,y,'ro')
    plt.plot(x1, poly(x1))
    plt.show()
```

```python
#The data points get further away from each other in the plot and the consistency and quality of the sample space 
#goes down stating that increasing the sample space may not always provide us with the best solution
```
